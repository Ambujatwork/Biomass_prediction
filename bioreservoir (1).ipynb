```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tools.sm_exceptions import ConvergenceWarning
import warnings


forecast_results = []  
for index, row in bio_his.iterrows():
    location = row['Latitude'], row['Longitude']
    time_series_data = row['2010':'2017']  # Extract data for 2010 to 2017


    time_series = pd.Series(time_series_data.values, index=pd.to_datetime(time_series_data.index, format='%Y'))

    try:
      
        order = (2, 1, 2)  # determine appropriate values for p, d, and q
        model = ARIMA(time_series, order=order)

        with warnings.catch_warnings():
            warnings.filterwarnings("ignore", category=ConvergenceWarning)
            model_fit = model.fit()

    
        next_years = pd.to_datetime(['2018', '2019'], format='%Y')
        forecast = model_fit.get_forecast(steps=2).predicted_mean.loc[next_years]

       
        forecast = forecast.clip(lower=0)

        forecast_results.append({
            
            'Forecast_2018': forecast['2018'],
            'Forecast_2019': forecast['2019']
        })
    except Exception as e:
        print(f"Error for {location}: {e}")

forecast_df = pd.DataFrame(forecast_results)

forecast_csv_path = 'forecasted_result.csv'
forecast_df.to_csv(forecast_csv_path, index=True)

print("Forecasted results saved to", forecast_csv_path)
```
